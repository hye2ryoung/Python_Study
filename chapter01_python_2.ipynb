```python
import pandas as pd
import numpy as np
```

```python
route1 = '/Users/kimhyeryoung/Documents/ds_study/Chapter 1. 공공 데이터를 통한 시장 동향 이해/Data/1_온라인쇼핑몰 운영형태별 상품군별거래액.csv'
```

```python
# 한글 인코딩
df = pd.read_csv(route1, encoding = 'cp949')

# melting

cate_cols = list(df.columns[:3])
result = pd.melt(df, id_vars = cate_cols, var_name = '날짜', value_name = '거래액')
```

```python
def strtoint(x) :

    if type(x) == str : #type이 string이면
        x = x.replace("-", "0") # - 를 0으로 바꿔줍니다 (참고. replace는 string에 적용되는 함수로, 바로 int로 바꿀 수는 없어요!)
        x = int(x) # int를 씌워서 int로 바꾼다

    else : #그 외의 type은 이미 Int이므로 pass 하겠습니다
        pass

    return x
```

```python
result["거래액"] = result["거래액"].apply(strtoint)

result = result[result["상품군별(1)"] != "합계"]
result = result[result["운영형태별(1)"] != "계"]
result = result[result["상품군별(2)"] == "소계"]
result = result.drop("상품군별(2)", axis = 'columns')
result['날짜'] = result['날짜'].apply(lambda x : x.replace(" p)", ""))
```

```python
result.to_csv("preprocess.csv", encoding = 'cp949', index = False)

#index = False 처리를 해주지 않으면 Unnamed 인덱스가 생겨난다
```

```python
for f in file_ls :
    print(f'{f}를 처리하고 있습니다')
    # 한글 인코딩
    df = pd.read_csv(os.path.join(src, f), encoding = 'cp949')

    # melting

    cate_cols = list(df.columns[:3])
    result = pd.melt(df, id_vars = cate_cols, var_name = '날짜', value_name = '거래액')

    #함수 적용
    result["거래액"] = result["거래액"].apply(strtoint)
    #합계빼기
    result = result[result[result.columns[0]] != "합계"]
    #중간집계빼기
    result = result[result[result.columns[2]] != '계']
    #가전,전자 / 통신기기 구분 빼기
    result = result[result[result.columns[1]] == '소계']
    #컬럼 삭제
    result = result.drop("상품군별(2)", axis = 'columns')

    # 추정치를 의미하는 'p)' 빼기
    result['날짜'] = result['날짜'].apply(lambda x : x.replace(" p)", ""))
    # 연도처리 쉽게 하도록 기호 바꾸기
    result['날짜'] = result['날짜'].apply(lambda x : x.replace(".", "/"))

    result.to_csv(os.path.join(dst, f), encoding = 'utf-8', index = False)
    print(f'{f} 완료')
```
