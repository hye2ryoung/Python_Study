#### Q57. Income_Category의 카테고리를 map 함수를 이용하여 다음과 같이 변경하여 newIncome컬럼에 매핑하라

Unknown : N

Less than $40K : a

$40K - $60K : b

$60K - $80K : c

$80K - $120K : d

$120K + : e

````python
dic = {
    'Unknown' : 'N',
    'Less than $40K' : 'a',
    '$40K - $60K' : 'b',
    '$60K - $80K' : 'c',
    '$80K - $120K' : 'd',
    '$120K +' : 'e'
}
df['newIncome'] = df.Income_Category.map(lambda x: dic[x])
#lambda x 사용자정의함수 Income_Category를 x로 취급
Ans = df[['newIncome','Income_Category']]
Ans.head()
````

#### Q58. Income_Category의 카테고리를 apply함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑

````python
def changeCategory(x):
  if x =='Unknown':
    return 'N'
  elif x =='Less than $40K':
    return 'a'
  elif x =='$40K - $60K':
    return 'b'
  elif x =='$60K - $80K':
    return 'c'
  elif x =='$80K - $120K':
    return 'd'
  elif x =='$120K +':
    return 'e'

df['newIncome'] = df.Income_Category.apply(changeCategory)

Ans = df[['newIncome','Income_Category']]
Ans.head()
````

#### Q76. 년-월-일:시 컬럼을 pandas에서 인식할 수 있는 datetime형태로 변경하라. 서울시의 제공데이터의 경우 0시가 24시로 표현된다.

````python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/파이썬기초/seoul_pm.csv')
df.head()
````

````python
def change_date(x):
  import datetime
  hour = x.split(':')[1] #시 데이터 (년-월-일:시)
  date = x.split(':')[0] #일 데이터

  if hour == '24':
    hour ='00:00:00'
    FinalDate = pd.to_datetime(date+" "+hour)+datetime.timedelta(days=1)
    # 24시 이후의 시간을 처리할때는 일수가 넘어가므로 timedelta(days=1)를 더해준다
  else:
    hour = hour+':00:00'
    FinalDate = pd.to_datetime(date+" "+hour)
  return FinalDate

df['(년-월-일:시)']=df['(년-월-일:시)'].apply(change_date)
Ans=df
Ans.head()
````
