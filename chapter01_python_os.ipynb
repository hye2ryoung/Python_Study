```python
for f in os.listdir(test_route) :
    print(f)
    print("파일명 : ", os.path.splitext(f)[0])
    print("타입 : ", os.path.splitext(f)[1])
    print("-------------------")
```

```python
for f in os.listdir(test_route) :
    if os.path.splitext(f)[1] == '.csv' :
        print(f)
        print("csv 입니다")
    else :
        print(f)
        print("csv가 아닙니다")
```
